```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_breast_cancer
#data=wdbc.dat
# Create a dataset with sample data
X, _ =load_breast_cancer(return_X_y=True)
#X = np.array([[1, 2], [1.5, 1.8], [5, 8], [8, 8], [1, 0.6], [9, 11]]
 #(n_samples=300, centers=2, cluster_std=0.60, random_state=0)

# Apply the K-means algorithm
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)

# Get the cluster centers
centroids = kmeans.cluster_centers_

# Predict the cluster for each data point
labels = kmeans.predict(X)

# Plot the clusters and their centroids
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X')  # Centroid points
plt.title('K-means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Print cluster centroids (dimensions)
print("Cluster Centroids (Dimensions):")
print(centroids)
```
